```python
from pipeline import Portfolio
import pandas as pd
```

```python
def compute_current(p):
    p.compute_weights()
    p.compute_numstocks()
    print(p.numstocks)
    
def bench(p, pstart, pend, benchstart, benchend):
    p.compute_weights(start = pstart, end = pend)
    p.compute_numstocks()

    date1 = p.stockdata.loc[(p.stockdata.index == benchstart) & (p.stockdata['Ticker'] != '^GSPC')].set_index('Ticker', drop=True)
    date1 = date1.loc[p.weights.index.values, 'Close']
    date2 = p.stockdata.loc[(p.stockdata.index == benchend) & (p.stockdata['Ticker'] != '^GSPC')].set_index('Ticker', drop=True)
    date2 = date2.loc[p.weights.index.values, 'Close']
    percchange = (date2 - date1) / date1
    print("portfolio: ", float(sum(percchange * p.weights)))
    
    spdate1 = p.stockdata.loc[(p.stockdata.index == '2021-05-01') & (p.stockdata['Ticker'] == '^GSPC')]['Close'].values
    spdate2 = p.stockdata.loc[(p.stockdata.index == '2021-07-01') & (p.stockdata['Ticker'] == '^GSPC')]['Close'].values
    print("market: ", float((spdate2 - spdate1) / spdate1))
```

```python
ticks = pd.read_csv("constituents_csv.csv")['Symbol'].tolist()
#ticks = ['AAPL', 'MSFT', 'TSLA']
p = Portfolio(tickers = ticks, value = 20000, start = '2016-05-01')
p.download()
p.clean_data(minmonths = 36)
```

```python
bench(p, min(p.stockdata.index), '2021-05-01', '2021-05-01', '2021-07-01')
```
